```python
import pandas as pd
print("Successfully imported")
```

```python
yearly=pd.read_csv("yearly_deaths_by_clinic.csv")
```

```python
#to read the first 10 rows from the dataset
yearly.head(10)
```

```python
#to print yearly
yearly
```

```python
#calculating the proportion of yearly deaths to yearly births and string with the column name proportion_deaths
yearly["proportion_deaths"]=(yearly['deaths']/yearly['births'])
```

```python
#proportion_deaths is added to the data frame
yearly.head(5)
```

```python
#Extract the rows from clinic 1 into yearly1 a
yearly1=yearly[yearly['clinic']=='clinic 1']
```

```python
yearly1
```

```python
#the rows from clinic 2 into yearly2.
yearly2=yearly[yearly["clinic"]=="clinic 2"]
```

```python
yearly2
```

```python
#Plot proportion_deaths by year for the two clinics in a single plot. Use the DataFrame plot method.
#Label the plotted lines using the label argument to plot.
#Save the Axes object returned by the plot method into the variable ax.
#Change the y-axis label to "Proportion deaths".
ax = yearly1.plot(x="year", y="proportion_deaths",
              label="yearly1")
yearly2.plot(x="year", y="proportion_deaths",
         label="yearly2",ax=ax)
ax.set_ylabel("Proportion deaths")
ax.set_xlabel("year")

#By capturing the ax object and giving it as an argument in the plot statement we get both lines in the same plot. The ax object can be used to further modify the plot. 
```

```python
#read the monthly_deaths csv
monthly=pd.read_csv("monthly_deaths.csv",parse_dates=["date"])
```

```python
monthly
```

```python
# finding the proportion_deaths by dividing the monthly[births] and monthly["deaths"]
monthly["proportion_deaths"]=(monthly["deaths"]/monthly["births"])
```

```python
monthly
```

```python
monthly.tail(25)
```

```python
#Plot proportion_deaths by date for the monthly date using the DataFrame plot method.
#Save the Axes object returned by the plot method into the variable ax.
#Change the y-axis label to "Proportion deaths"
ax = monthly.plot(x="date", y="proportion_deaths",
              label="monthly")
ax.set_ylabel("Proportion deaths")
```

```python
#from the graph we can make out that after 1847 the death rates is being reduced drastically so we take the date in year 1847
handwashing_start = pd.to_datetime('1847-06-01')
```

```python
handwashing_start
```

```python
#record at the year 1847
at_washing = monthly[
  monthly["date"] == handwashing_start]
```

```python
at_washing
```

```python
#records before handwashing
before_washing = monthly[
  monthly["date"] < handwashing_start]
```

```python
before_washing
```

```python
#records after hand washing
after_washing = monthly[
  monthly["date"] >= handwashing_start]
```

```python
after_washing
```

```python
#plotting the graph before handwashing and after handwashing
ax = before_washing.plot(x="date", y="proportion_deaths",
              label="before_washing")
after_washing.plot(x="date", y="proportion_deaths",
         label="after_washing", ax=ax)
ax.set_ylabel("Proportion deaths")
```

```python
#Select the column proportion_deaths in before_washing and put it into before_proportion
before_proportion=before_washing['proportion_deaths']
```

```python
before_proportion
```

```python
#Do the same for proportion_deaths in after_washing and put it into after_proportion.
after_proportion=after_washing['proportion_deaths']
```

```python
after_proportion
```

```python
#mean or the avg of before_proportion
before_proportion.mean()
```

```python
#mean of after proportion
after_proportion.mean()
```

```python
#mean difference between after_proportion and before_proportion
mean_diff=after_proportion.mean()-before_proportion.mean()
```

```python
mean_diff
```

```python
boot_before = before_proportion.sample(frac=1, replace=True)
```

```python
boot_before
```

```python
boot_after = after_proportion.sample(frac=1, replace=True)
```

```python
boot_after
```

```python
boot_mean_diff = []
for i in range(3000):
    boot_before = before_proportion.sample(frac=1, replace=True)
    #print(boot_before)
    boot_after = after_proportion.sample(frac=1, replace=True)
    boot_mean_diff.append(boot_after.mean()-boot_before.mean())
```

```python
boot_mean_diff
```

```python
confidence_interval1=pd.Series(boot_mean_diff)
confidence_interval1
```

```python
confidence_interval=pd.Series(boot_mean_diff).quantile([0.025, 0.975])
```
